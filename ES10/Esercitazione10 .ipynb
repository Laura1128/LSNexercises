### Exercise 10.1

L'obiettivo di questo esercizio è risolvere il problema del commesso viaggiatore utilizzando un algoritmo di simulated annealing. Partendo da un singolo individuo, vengono proposte delle mosse(mutazioni), che possono essere accettate o rifiutate. L'accettazione o il rifiuto della mossa dipendono dal fitness dell'individuo considerato e dell'individuo mutato, oltre che dalla temperatura fittizia assegnata. Si è scelto un valore di temperatura iniziale 
$T_{i}=1.25$, per il quale si hanno valori di accettazione abbastanza alti (70-80%). La temperatura viene quindi progressivamente abbassata di $\delta T=0.02$  fino ad arrivare al valore $T_{f}=0.01$.

Il fitness dell'individuo presenta grosse oscillazioni nelle prime iterazioni, che poi vanno via via riducendosi con l'abbassarsi della temperatura. Nelle prime fasi infatti l'accettazione è molto alta e questo fa sì che possano essere esplorate diverse configurazioni. Con l'abbassarsi della temperatura l'accettazioni diventa sempre piu bassa, costringendo l'algoritmo ad assestarsi su valori di fitness sempre più bassi.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x, y=np.loadtxt("SimAnn/BestPath1.dat", unpack=True, usecols = (0,1)) 
x_start=-0.754701 
y_start=0.430426 
x1, y1=np.loadtxt("SimAnn/BestPath2.dat", unpack=True, usecols = (0,1)) 
x1_start=1.22649356722832e-01   
y1_start=1.69917970895767e-01
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(9, 4))
fig.suptitle('Simulated Annealing: Best Paths')
ax1.plot(x, y,marker='o', color='blue')
ax1.plot(x_start, y_start, color="red", marker = "o")
ax1.set_title('Cities on a circumference')
ax1.set(xlabel='x', ylabel='y')
ax1.grid(True)
ax2.plot(x1, y1,  marker='o', color='green')
ax2.plot(x1_start, y1_start, color="red", marker = "o")
ax2.set_title('Cities in a square')
ax1.set(xlabel='x', ylabel='y')
ax2.grid(True)

plt.show()
```

```python
x1, data1=np.loadtxt("SimAnn/CircleMetropolis.dat", unpack=True, usecols = (0,1)) 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

fig.suptitle('Cities on a circumference')

ax1.plot(x1, data1)
ax1.set(xlabel='#iterazioni', ylabel='L')
ax1.set_title('Simulated Annealing : Fitness vs iterazioni')
ax1.grid(True)

x2, data2=np.loadtxt("SimAnn/risultatiGA/BestFitCircle.dat", unpack=True, usecols = (0,1)) 
ax2.plot(x2, data2)
ax2.set(xlabel='#generazione', ylabel='L')
ax2.set_title('Genetic Algorithm: Fitness vs generazioni')
ax2.grid(True)

plt.show()
```

### Exercise 10.2

In questo esercizio la ricerca del percorso migliore viene portata avanti parallelamente da 4 nodi di calcolo a partire dalla stessa generazione.

Ogni 50 generazioni ogni nodo manda agli altri il proprio migior individuo tramite una comunicazione allgather. I primi 4 individui di ogni nodo vengono così rimpiazzati con i quattro individui migliori dei quattro nodi.

Ogni nodo trova lo stesso miglior percorso finale.



```python
x, y=np.loadtxt("MessagePassing/BestPath0.dat", unpack=True, usecols = (0,1)) 
x_start=-0.754701 
y_start=0.430426 
x1, y1=np.loadtxt("MessagePassing/BestPath1.dat", unpack=True, usecols = (0,1))
x2, y2=np.loadtxt("MessagePassing/BestPath2.dat", unpack=True, usecols = (0,1)) 
x3, y3=np.loadtxt("MessagePassing/BestPath3.dat", unpack=True, usecols = (0,1)) 

fig, (ax1, ax2, ax3, ax4) = plt.subplots(1, 4, figsize=(16, 4))
fig.suptitle('Message Passing: Best Paths')
ax1.plot(x, y,marker='o', color='blue')
ax1.plot(x_start, y_start, color="red", marker = "o")
ax1.set_title('Nodo 0')
ax1.set(xlabel='x', ylabel='y')
ax1.grid(True)
ax2.plot(x1, y1,  marker='o', color='blue')
ax2.plot(x_start, y_start, color="red", marker = "o")
ax2.set_title('Nodo 1')
ax2.set(xlabel='x', ylabel='y')
ax2.grid(True)
ax3.plot(x2, y2,  marker='o', color='blue')
ax3.plot(x_start, y_start, color="red", marker = "o")
ax3.set_title('Nodo 2')
ax3.set(xlabel='x', ylabel='y')
ax3.grid(True)
ax4.plot(x3, y3,  marker='o', color='blue')
ax4.plot(x_start, y_start, color="red", marker = "o")
ax4.set_title('Nodo3')
ax4.set(xlabel='x', ylabel='y')
ax4.grid(True)

plt.show()
```
