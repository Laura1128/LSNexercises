Il codice utilizzato in questa esercitazione permette di simulare un modello di Ising 1D con metodi Monte Carlo.

Con gli algoritmi di Metropolis e di Gibbs si possono facilmente campionare distribuzioni di probabilità date dal peso statistico di Boltzmann $p(\{ s_i \}) = \exp(-\beta H)/Z$. Tuttavia occorre tenere presente che il campionamento è corretto solo asintoticamente e che le configurazioni di spin visitate in step temporali vicini sono altamente correlate. Per ovviare al primo problema è sufficiente far evolvere il sistema per un breve periodo fino al raggiungimento dell'equilibro.  Per far fronte al secondo problema esistono diverse alternative, una delle quali è la media a blocchi. Seguendo la curva dell'incertezza statistica all'aumentare della lunghezza dei blocchi L si osserva che dopo una rapida crescita l'incertezza satura ad un valore limite. Il valore di L per cui l'incertezza smette di crescere corrisponde alla lunghezza minima che garantisce l'indipendenza statistica tra i valori medi delle osservabili misurate. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

x, data=np.loadtxt("Energy_incertezza.dat", unpack=True, usecols = (0,1)) 
ax1.plot(x,data)
ax1.set_title('incertezza su U')
ax1.set(xlabel='L', ylabel='incertezza')
ax1.grid(True)

x2, data2=np.loadtxt("Heat_incertezza.dat", unpack=True, usecols = (0,1)) 
ax2.plot(x2,data2)
ax2.set_title('incertezza su C')
ax2.set(xlabel='L', ylabel='incertezza')
ax2.grid(True)
plt.show()

```

Sulla base dei grafici ottenuti scelgo L=1000 come lunghezza dei blocchi.

Nei grafici seguenti sono mostrati i risultati ottenuti con l'Algoritmo di Metropolis (in giallo) e Gibbs(in rosso) facendo 19 misure di:
1. $U(N,T) = \langle H \rangle_{h=0}$
2. $C(N,T) = k_B \beta^2 (\langle H^2 \rangle_{h=0}-\langle H \rangle_{h=0}^2)$ 
3. $\chi(N,T) = \beta \langle (\sum_{i=1,N}s_i)^2 \rangle_{h=0}$
4. $M(N,T,h=0.02)= \langle \sum_{i=1,N}s_i \rangle_{h=0.02}$

a T compresa fra 0.5 e 2 dopo aver equilibrato il sistema. In blu per confronto la curva trovata con il metodo analitico.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pylab

points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )

x1, data1, error1=np.loadtxt("Energy0.dat", unpack=True, usecols = (0,1,2))
x2, data2, error2=np.loadtxt("Energy0Gibbs.dat", unpack=True, usecols = (0,1,2))

pylab.figure(figsize=(11,6))
pylab.plot(T, e, '-b', label='soluzione esatta')
pylab.errorbar(x1, data1, yerr=error1, color='y', marker='s', label='Metropolis')
pylab.errorbar(x2, data2, yerr=error2, color='r', marker='o', label='Gibbs')

pylab.legend(loc='upper left')

plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')
plt.grid(True)

plt.show()
```

```python
heat=((beta*J)**2)*(((1+thN+(Ns-1)*(th**2)+(Ns-1)*(ch**2)*thN)/(1+thN))-Ns*((th+ch*thN)/(1+thN))**2)
x1, data1, error1=np.loadtxt("Heat0.dat", unpack=True, usecols = (0,1,2))
x2, data2, error2=np.loadtxt("Heat0Gibbs.dat", unpack=True, usecols = (0,1,2))

pylab.figure(figsize=(11,6))
pylab.plot(T, heat, '-b', label='soluzione esatta')
pylab.errorbar(x1, data1, yerr=error1, color='y', marker='s', label='Metropolis')
pylab.errorbar(x2, data2, yerr=error2, color='r', marker='o', label='Gibbs')

pylab.legend(loc='upper right')

plt.title('Ising 1D,  Heat capacity')
plt.xlabel('T')
plt.ylabel('C')
plt.grid(True)

plt.show()
```

```python
X = beta*np.exp(2*beta*J)*(1-thN)/(1+thN)
x1, data1, error1=np.loadtxt("Chi0.dat", unpack=True, usecols = (0,1,2))
x2, data2, error2=np.loadtxt("Chi0Gibbs.dat", unpack=True, usecols = (0,1,2))

pylab.figure(figsize=(11,6))
pylab.plot(T, X, '-b', label='soluzione esatta')
pylab.errorbar(x1, data1, yerr=error1, color='y', marker='s', label='Metropolis')
pylab.errorbar(x2, data2, yerr=error2, color='r', marker='o',  label='Gibbs')

pylab.legend(loc='upper right')

plt.title('Ising 1D, Susceptibility')
plt.xlabel('T')
plt.ylabel('$\chi$')
plt.grid(True)

plt.show()
```

```python
h=0.02 #external field
b = 1/T

l1 = np.exp(b*J)*np.cosh(b*h)+np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
l2 = np.exp(b*J)*np.cosh(b*h)-np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))
Z = l1**Ns + l2**Ns
M = (np.exp(b*J)*np.sinh(b*h)*((l1**(Ns-1))*(1+np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J))) 
        + (l2**(Ns-1))*(1-np.exp(b*J)*np.cosh(b*h)/np.sqrt(np.exp(2*b*J)*np.cosh(b*h)*np.cosh(b*h)-2*np.sinh(2*b*J)))))/(Z)

x1, data1, error1=np.loadtxt("Mag.dat", unpack=True, usecols = (0,1,2))

x2, data2, error2=np.loadtxt("MagGibbs.dat", unpack=True, usecols = (0,1,2))


pylab.figure(figsize=(11,6))
pylab.plot(T, M, '-b', label='soluzione esatta')
pylab.errorbar(x1, data1, yerr=error1, color='y', marker='s', label='Metropolis')
pylab.errorbar(x2, data2, yerr=error2, color='r', marker='o',  label='Gibbs')

pylab.legend(loc='upper right')
plt.title('Ising 1D, magnetization M with h = 0.02')
plt.xlabel('T')
plt.ylabel('$M$')
plt.grid(True)

plt.show()
```
