```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


N=1000                
x = np.arange(N) 
data=np.loadtxt("equilibrazione/raggio_istantaneo1s.dat") 
plt.plot(x,data)
y = 0*x + 1.5
plt.plot(x,y, color='r')
plt.xlabel('#throws')
plt.ylabel('<r/a>')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =np.loadtxt("equilibrazione/stato1s.dat", unpack=True, usecols = (0,1,2), delimiter=' ')
fig=plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.view_init(10,30)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=1000000
N=1000                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N) 
#err = np.arange(N)
data, error=np.loadtxt("misure1s.dat", unpack=True, usecols = (0,1), delimiter=' ') 
x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,data, yerr=error)
y = 0*x + 1.5
plt.plot(x,y, color='r')
plt.xlabel('#throws')
plt.ylabel('<r/a>')
plt.grid(True)
plt.show()
```
