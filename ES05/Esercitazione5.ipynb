### Exercise 05.1

In meccanica quantistica la funzione d'onda $\Psi$ rappresenta lo stato di un sistema fisico e può essere interpretata come ampiezza di probabilità. Detta $\Psi(\vec{r},t)$ la funzione d'onda di una particella non relativistica, il modulo quadro $|\Psi(\vec{r},t)|^2$ rappresenta la densità di probabilità di trovare la particella nel punto $\vec{r}$ al tempo $t$.

Utilizzando l'algoritmo di Metropolis si vuole campionare il modulo quadro della funzione d'onda dello stato fondamentale 1s e dello stato eccitato 2p dell'atomo d'idrogeno $|\Psi_{1,0,0}(x,y,z)|^2$ e $|\Psi_{2,1,0}(x,y,z)|^2$.

Funzione d'onda dello stato fondamentale 1s:
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
Funzione d'onda dello stato eccitato 2p:
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$


Con le densità di probabilità campionate sarà poi possibile stimare il valor medio del raggio nei due stati: $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ e $\left\langle r \right\rangle_{\Psi_{2,1,0}}$

Valori attesi:
$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

<span style="color:red">Accettazione

È importante scegliere un passo della matrice di trasferimento stocastica tentativo $T(\vec{x}|\vec{y})$ tale che l'accettazione sia intorno al 50%. In questo modo si raggiunge il giusto compromesso fra scorrelazione tra i dati e efficienza computazionale.

Di seguito è mostrato un esempio di misure effettuate con un passo che da un'accettazione superiore all'80%

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =np.loadtxt("Accettazione/stato1s.dat", unpack=True, usecols = (0,1,2), delimiter=' ')
fig=plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.view_init(10,30)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z =np.loadtxt("Accettazione/stato2p.dat", unpack=True, usecols = (0,1,2), delimiter=' ')
fig=plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.view_init(10,30)
plt.show()
```

```python
M=1000000
N=1000                
L=int(M/N)            
x = np.arange(L) 
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
data, error=np.loadtxt("Accettazione/misure1s.dat", unpack=True, usecols = (0,1), delimiter=' ') 
data2, error2=np.loadtxt("Accettazione/misure2p.dat", unpack=True, usecols = (0,1), delimiter=' ') 
x*=N 
ax1.errorbar(x,data, yerr=error)
ax1.axhline(y=1.5, color='r')
ax1.set_title('stato 1s')
ax1.set(xlabel='#throws', ylabel='<r/a>')
ax1.grid(True)
ax2.errorbar(x,data2, yerr=error2)
ax2.axhline(y=5, color='r')
ax2.set_title('stato 2p')
ax2.set(xlabel='#throws', ylabel='<r/a>')
ax2.grid(True)

plt.show()
```

L'utilizzo di un passo molto piccolo fa sì che la grande maggioranza delle mosse vengano accettate. In questo modo lo spazio delle configurazioni viene esplorato molto lentamente e le configurazioni presentano un alto grado di correlazione.

Per poter avere un'accettazione intorno al 50% è stato scelto un passo pari ad $a$ per lo stato $1s$ e un passo pari a $3a$ per lo stato $2p$.
