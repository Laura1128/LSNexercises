### Exercise 07.1

Dopo aver fatto compiere al sistema 500 passi, in  modo tale che raggiunga l'equilibrio, stampo 500000 valori istantanei di energia potenziale per particella e pressione. Con questi dati calcolo l'autocorrelazione per poter fare una prima stima della lunghezza dei blocchi da usare nel calcolo dei valori medi e delle incertezze.

- fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ ($r_c = 2.2$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

Epot=np.loadtxt("es7.1/solid/instant_Epot")
Pres=np.loadtxt("es7.1/solid/instant_P")


def funzione_autocorrelazione(array, Lag):
    autocorr = np.zeros(Lag)
    tmax=500000
    for i in range(Lag):
        v=0
        for l in range(tmax - i):
            v+=array[l]*array[l+i]
        v=v/(tmax-i)
        
        b=0
        d=0
        for l in range(tmax - i):
            b+=array[l]
            d+=array[l+i]
        f=b/(tmax-i) * d/(tmax-i)
        
        q=0
        r=0
        for l in range(tmax):
            q+=array[l]*array[l]
            r+=array[l]
        q=q/(tmax)
        r=(r/(tmax))**2
         
        autocorr[i] = (v - f)/(q - r)
    return autocorr;
 
A=funzione_autocorrelazione(Epot, 300)
B=funzione_autocorrelazione(Pres, 300)
x = np.arange(300)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))
ax1.plot(x,A)
ax2.plot(x,B)
ax1.set_title('autocorrelazione di U/N')
ax1.set(xlabel='time step', ylabel='autocorrelazione')
ax2.set_title('autocorrelazione di P')
ax2.set(xlabel='time step', ylabel='autocorrelazione')
ax1.grid(True)
ax2.grid(True)

plt.show()
```

Dai grafici risulta che l'autocorrelazione, per entrambe le quantità misurate, decade dopo un centinaio di passi. La lunghezza dei blocchi dovrà essere comunque molto piu grande del tempo di decadimento dell'autocorrelazione; per fare un'analisi piu precisa studio il variare dell'incertezza all'aumentare della lunghezza dei blocchi impiegati per calcolare le medie dei valori misurati.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=500000              # Total number of throws

Epot=np.loadtxt("es7.1/solid/instant_Epot")

Nstep=[50000,25000,10000,5000,4000,2500,1000,800,500,250,200,100]                 # Number of blocks
vec_err=np.zeros(12)
for h in range(12):
    N=Nstep[h]
    L=int(M/N)            
    ave = 0
    av2 = 0
    sum_prog = 0
    su2_prog=0
    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += Epot[k]
        ave = sum/L       # r_i 
        av2 = (ave)**2 # (r_i)^2 
        sum_prog += ave
        su2_prog += av2

    sum_prog=sum_prog/N
    su2_prog=su2_prog/N
    err = math.sqrt((su2_prog - sum_prog**2)/N) # Statistical uncertainty
    vec_err[h]=err 


x=[10,20,50,100,125,200,500,625,1000,2000,2500,5000]   #lunghezza blocchi
plt.plot(x, vec_err)
plt.title('Energia potenziale')
plt.xlabel('L')
plt.ylabel('incertezza')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=500000              # Total number of throws

Pres=np.loadtxt("es7.1/solid/instant_P")

Nstep=[50000,25000,10000,5000,4000,2500,1000,800,500,250,200,100]                 # Number of blocks
vec_err=np.zeros(12)
for h in range(12):
    N=Nstep[h]
    L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
    ave = 0
    av2 = 0
    sum_prog = 0
    su2_prog=0
    for i in range(N):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += Pres[k]
        ave = sum/L       # r_i 
        av2 = (ave)**2 # (r_i)^2 
        sum_prog += ave
        su2_prog += av2

    sum_prog=sum_prog/N
    su2_prog=su2_prog/N
    err = math.sqrt((su2_prog - sum_prog**2)/N) # Statistical uncertainty
    vec_err[h]=err 

x=[10,20,50,100,125,200,500,625,1000,2000,2500,5000] # Number of throws in each block
plt.plot(x, vec_err)
plt.title('Pressione')
plt.xlabel('L')
plt.ylabel('incertezza')
plt.grid(True)
plt.show()
```

Si può notare che dopo una rapida crescita iniziale l'incertezza varia molto lentamente. Una valore di L=1000 sembra dunque essere una valida scelta per poter avere dati scorrelati.

### Exercise 07.4

La funzione di distribuzione radiale $g(r)$ dà la probabilità di trovare una coppia di atomi a distanza $r$ l'uno dall'altro, rispetto alla probabilità di trovare la stessa coppia di atomi a quella distanza in un gas ideale. Questa fornisce informazioni su come si distribuiscono gli atomi in certe condizioni e permette di calcolare il valor medio di funzioni che dipendono dalle coordinate di coppie di particelle.

Di seguito vengono confrontate le funzioni di distribuzioni radiali calcolate con il codice Monte Carlo NVT e con il codice di dinamica molecolare NVE. Vengono inoltre riportati i valori medi di energia potenziale per particella e pressione calcolati con il metodo a blocchi. I parametri utilizzati sono quelli dell'Argon (($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu).

1. solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

```python
import pylab

k=1.3806488*10**(-23)
epsilon=120*k 
sigma=0.34*10**(-9)

raggio, data, error=np.loadtxt("es7.4/MonteCarlo_NVTupgraded/solid/output.gave.0", unpack=True, usecols = (0,1,2))
raggio2, data2, error2=np.loadtxt("es7.4/MolDyn/solid/output.gave.0", unpack=True, usecols = (0,1,2))
raggio=raggio*sigma
raggio2=raggio2*sigma
pylab.figure(figsize=(11,6))
pylab.errorbar(raggio, data, yerr=error, color='b', label='g(r) MC_NVT')
pylab.errorbar(raggio2, data2, yerr=error2, color='g', label='g(r) MolDyn_NVE')
pylab.legend(loc='upper right')

plt.xlabel('r[m]')
plt.ylabel('g(r)')
plt.grid(True)


plt.show()
```

```python
M=200000
L=1000              #lunghezza del blocco
N=200                # Number of blocks

k=1.3806488*10**(-23)
epsilon=120*k 
sigma=0.34*10**(-9)


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 4))
x = np.arange(N) 
data, error=np.loadtxt("es7.4/MonteCarlo_NVTupgraded/solid/output.epot.0", unpack=True, usecols = (2,3))
data1, error1=np.loadtxt("es7.4/MonteCarlo_NVTupgraded/solid/output.pres.0", unpack=True, usecols = (2,3))
data*=epsilon 
error*=epsilon
data1*=epsilon/sigma**(3)
error1*=epsilon/sigma**(3)
ax1.errorbar(x,data, yerr=error)
ax2.errorbar(x,data1, yerr=error1)
ax1.set_title('Energia potenziale per particella')
ax1.set(xlabel='#blocchi', ylabel='<U/N> [J]')
ax2.set_title('Pressione')
ax2.set(xlabel='#blocchi', ylabel='<P> [Pa]')
plt.show()
```
