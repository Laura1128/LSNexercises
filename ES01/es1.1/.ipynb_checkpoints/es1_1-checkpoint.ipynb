```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N) 
#err = np.arange(N)
myfile=open("medie.txt", "r")
file=open("incertezze.txt", "r")
medie=myfile.readlines()
incertezze=file.readlines()
for i in range(0,len(medie)):
    medie[i]=float(medie[i].strip("\n"))
    incertezze[i]=float(incertezze[i].strip("\n"))

arraym=np.array(medie)
arrayi=np.array(incertezze)

myfile.close()
    
x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,arraym-1/12, yerr=arrayi)
plt.xlabel('#throws')
plt.ylabel('<r>')
plt.grid(True)
plt.show()
```
