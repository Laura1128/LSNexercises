### Exercise 01.1

L'obiettivo di questo esercizio è testare il generatore di numeri pseudo-casuali. A tal fine possono essere condotti 3 test:
- Calcolare il valor medio della distribuzione uniforme tra 0 e 1, verificando che $\langle r \rangle = \int_0^1 r dr = 1/2$. 
- Calcolare la varianza della distribuzione uniforme fra 0 e 1, verificando che $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$.
- Test del $\chi^2$.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block
x = np.arange(N) 

data=np.loadtxt("es1.0/medie.txt")
error=np.loadtxt("es1.0/incertezze.txt")  

plt.errorbar(x,data-0.5, yerr=error)
plt.xlabel('N')
plt.ylabel('<r>-1/2')
plt.grid(True)
plt.show()
```

Nel grafico è mostrato l'andamento del valore di $\langle r \rangle - 1/2$ e della sua incertezza calcolata con il metodo a blocchi.


```python
M=100000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block
x = np.arange(N) 

data=np.loadtxt("es1.1/medie.txt")
error=np.loadtxt("es1.1/incertezze.txt")  

plt.errorbar(x,data-1/12, yerr=error)
plt.xlabel('N')
plt.ylabel('<$\sigma^{2}$>')
plt.grid(True)
plt.show()
```

Nel grafico è mostrato l'andamento di $\sigma^2-1/2$ e della sua incertezza in funzione del numero di blocchi $N$.

Supponiamo di suddividere l'intervallo [0,1] in $M$ sottointervalli identici. Se i numeri sono distribuiti uniformemente, in ogni sottointervallo ci si aspetta di trovare la stessa quantità di numeri. Per verificare che i numeri estratti con il generatore siano effettivamente distribuiti uniformemente implemento il test del $\chi^2$, che si basa sulla misura della discrepanza tra le frequenze osservate $O_i$ e quelle attese $E_i$:
$$\chi^2 = \sum_{i} \frac{\left( O_i - E_i \right)^2}{E_i}$$
La probabilità di ottenere $k$ successi indipendenti, ognuno con probabilità $p$, in $n$ tentativi, è data dalla distribuzione binomiale:
$$Pr(X=k) = {{n}\choose{k}} p^k (1-p)^{n-k} = \frac{n!}{k! (n-k)!} p^k (1-p)^{n-k}$$
caratterizzata da valor medio $\langle X \rangle = np$ e varianza $\sigma^2 = np (1-p)$.
Se $p$ è piccola (in questo caso se suddivido in tanti sottointervalli) si ha che  $\sigma^2 \simeq np$, quindi $\sigma^2 \simeq \langle X \rangle$. Questo spiega perchè al denominatore nella formula del $\chi^2$ compare $E_i$.


Nel caso in questione $$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$ done $n_i$ rappresenta la quantità di numeri appartenenti all'intervallo i-esimo mentre $n$ è la quantità totale di numeri estratti.

Fissato $M=100$, il primo valore di $\chi^2$ verrà calcolato considerando i primi $n=10^4$ numeri pseudocasuali, il secondo i successivi $n=10^4$ numeri pseudocasuali e così via fino ad ottenere 100 valori di $\chi^2$. Dato che in media deve essere $(n_i - n/M)^2 \simeq n/M$, segue che $\chi^2 \simeq 100$, cioè il nunero di sotto-intervalli.

```python
x = np.arange(N) 
data=np.loadtxt("es1.2/chiquadrati.txt") 
plt.figure(figsize=(10,5))
plt.plot(x,data, marker ='o')
plt.xlabel('i')
plt.ylabel(r'$\chi^2$')
plt.title('Test del chi quadrato')
plt.grid(True)
plt.show()

print("Media:", data.mean())
```

Le stime del $\chi^2$ oscillano attorno al valore atteso. Questo indica che l'ipotesi che i numeri pseudocasuali siano distribuiti uniformente non è improbabile.

### Exercise 01.2

Il teorema del limite centrale afferma che la distribuzione della somma (o della media) di un gran numero di variabili casuali indipendenti e identicamente distribuite tende a distribuirsi normalmente, indipendentemente dalle distribuzioni delle singole variabili. Questo teorema può essere generalizzato al caso di variabili che non sono identicamente distribuite, purchè tutte le variabili abbiano varianza finita.

In questo esercizio si vuole verificare il teorema del limite centrale costruendo gli istogrammi ottenuti con $10^4$ realizzazioni di $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (per $N=1, 2, 10, 100$), dove $x_i$ è una variabile che segue:
* una distribuzione uniforme
* una distribuzione esponenziale ($\lambda=1$)
* una distribuzione di Cauchy-Lorentz ($\mu=0$ and $\Gamma=1$)


```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 9))
fig.suptitle('Distribuzione uniforme')

datauni=np.loadtxt("es2/uniforme.txt")
n_bin=100
n, bins, patches = ax1.hist(datauni,n_bin,density=True,range=(0,1))
ax1.set_title("N=1")

datauniA=np.loadtxt("es2/uniformeA.txt")
n, bins, patches = ax2.hist(datauniA,n_bin,density=True,range=(0,1))
ax2.set_title("N=2")

datauniB=np.loadtxt("es2/uniformeB.txt")
n, bins, patches = ax3.hist(datauniB,n_bin,density=True,range=(0,1))
ax3.set_title("N=10")

datauniC=np.loadtxt("es2/uniformeC.txt")
n, bins, patches = ax4.hist(datauniC,n_bin,density=True,range=(0,1))
ax4.set_title("N=100")
mu=0.5
sigma=0.03
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax4.plot(bins, y, '--')

plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 9))
fig.suptitle('Distribuzione esponenziale')

datauni=np.loadtxt("es2/esponenziale.txt")
n_bin=100
n, bins, patches = ax1.hist(datauni,n_bin,density=True,range=(0,2))
ax1.set_title("N=1")

datauniA=np.loadtxt("es2/esponenzialeA.txt")
n, bins, patches = ax2.hist(datauniA,n_bin,density=True,range=(0,2))
ax2.set_title("N=2")

datauniB=np.loadtxt("es2/esponenzialeB.txt")
n, bins, patches = ax3.hist(datauniB,n_bin,density=True,range=(0,2))
ax3.set_title("N=10")

datauniC=np.loadtxt("es2/esponenzialeC.txt")
n, bins, patches = ax4.hist(datauniC,n_bin,density=True,range=(0,2))
ax4.set_title("N=100")

mu=1.0
sigma=0.1
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax4.plot(bins, y, '--')


plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 9))
fig.suptitle('Distribuzione di Cauchy-Lorentz')

datauni=np.loadtxt("es2/lorenziana.txt")
n_bin=100
n, bins, patches = ax1.hist(datauni,n_bin,density=True,range=(-4,4))
ax1.set_title("N=1")

datauniA=np.loadtxt("es2/lorenzianaA.txt")
n, bins, patches = ax2.hist(datauniA,n_bin,density=True,range=(-4,4))
ax2.set_title("N=2")

datauniB=np.loadtxt("es2/lorenzianaB.txt")
n, bins, patches = ax3.hist(datauniB,n_bin,density=True,range=(-4,4))
ax3.set_title("N=10")

datauniC=np.loadtxt("es2/lorenzianaC.txt")
n, bins, patches = ax4.hist(datauniC,n_bin,density=True,range=(-4,4))
ax4.set_title("N=100")
gamma=1.0
y = ((gamma/ np.pi) *(1 / (bins**2+ gamma**2)))
ax4.plot(bins, y, '--')

plt.show()
```

Osservando i grafici si nota che nei primi due casi la distribuzione della somma per N=100 può essere approssimata da una gaussiana, mentre nel terzo caso segue la distribuzione delle variabili di partenza, cioè la lorenziana. Questo significa che il teorema del limite centrale è verificato per le distribuzioni uniforme ed esponenziale, mentre non lo è per la distribuzione lorenziana. Infatti la distribuzione lorenziana ha varianza non finita e dunque non soddisfa le ipotesi del teorema del limite centrale.

### Exercise 01.3

Un ago di lunghezza L è lanciato su un piano orizzontale con righe parallele a distanza $d$ l'una dall'altra. La probabilità che l'ago intersechi una di queste linee risulta pari a  $P = 2L/\pi d$. È dunque possibile stimare $\pi$ a partire dalla probabilità $P$: $$\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$$
Detto $\phi$ l'angolo che l'ago forma con le rette e $x$ la distanza del centro dell'ago dalla retta più vicina, l'ago interseca una delle linee se è soddisfatta la condizione $x < \frac{l}{2}sin(\phi) $, con $0 < x < d/2, 0 < \phi < \pi$. Dunque si può calcolare $P$ contando il numero di volte in cui $x<\frac{l}{2}sin(\theta)$, dove $x$ è un numero uniformemente distribuito fra 0 e $d/2$, mentre $\theta$ è un angolo distribuito uniformemente fra 0 e $\pi$.

```python
from math import pi
M=10000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block
x = np.arange(N) 

data=np.loadtxt("Buffon/medie.txt")
error=np.loadtxt("Buffon/incertezze.txt")  
x*=L # Number of throws = block * (Number of throws in each block)
plt.errorbar(x,data, yerr=error)
plt.axhline(y=pi, color='red')

plt.xlabel('#throws')
plt.ylabel('<$\pi$>')
plt.grid(True)
plt.show()
```

Per campionare uniformemente un angolo fra 0 e $\pi$ si può utilizzare una tecnica di rigetto: vengono estratti due numeri x e y, il primo distribuito uniformemente fra -1 e 1, il secondo distribuito uniformemente fra 0 e 1. Se $x^{2} + y^{2} <1$ viene generato l'angolo $\theta= cos^{-1}(\frac{x}{\sqrt{x^2 + y^2}})$, altrimenti viene reiterato il procedimento finchè i numeri estratti non soddisfano la condizione posta.
