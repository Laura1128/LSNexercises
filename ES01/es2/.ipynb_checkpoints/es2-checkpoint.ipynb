```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

datauni=np.loadtxt("uniforme.txt")
n_bin=100
n, bins, patches = plt.hist(datauni,n_bin,range=(0,1))
plt.figure()
datauniA=np.loadtxt("uniformeA.txt")
n_bin=100
n, bins, patches = plt.hist(datauniA,n_bin,range=(0,1))
plt.figure()
datauniB=np.loadtxt("uniformeB.txt")
n_bin=100
n, bins, patches = plt.hist(datauniB,n_bin,range=(0,1))
plt.figure()
datauniC=np.loadtxt("uniformeC.txt")
n_bin=100
n, bins, patches = plt.hist(datauniC,n_bin,range=(0,1))

plt.grid(True)
plt.show()
```
