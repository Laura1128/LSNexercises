```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=100000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block, please use for M a multiple of N
x = np.arange(N) 
#err = np.arange(N)
myfile=open("chiquadrati.txt", "r")
chi=myfile.readlines()
for i in range(0,len(chi)):
    chi[i]=float(chi[i].strip("\n"))

arraychi=np.array(chi)

myfile.close()
    
#x*=L # Number of throws = block * (Number of throws in each block)
plt.plot(x,arraychi)
plt.xlabel('j')
plt.ylabel('chi^2')
plt.title('Test del chi quadrato')
plt.grid(True)
plt.show()
```
