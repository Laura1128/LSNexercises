### The Traveling Salesman Problem

### Exercise 09.1

L'algoritmo genetico sviluppato per questa esercitazione funziona nel seguente modo:
- crea una popolazione di 100 individui a partire dal primo generato, applicando 4 diverse mutazioni con una probabilità del 30% l'una.
- ordina la popolazione in base al fitness, calcolato misurando la distanza totale percorsa $L^{(1)}$
- seleziona due individui con un algoritmo che privilegia gli individui con fitness migliore: $j=N\times int(r^2)$ dove $r$ è un numnero casuale $r\in[0,1)$
- chiama uno alla volta i quattro operatori di mutazione, che intervengono con una probabilità del 10% l'uno
- viene chiamato l'operatore di crossover, che interviene con l'80% di probabilità 

Il processo viene reiterato fino a produrre 92 individui, che si aggiungeranno agli otto migliori della generazione precedente per costituire la nuova generazione.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pylab

x1, data1=np.loadtxt("risultati/BestFitCircle.dat", unpack=True, usecols = (0,1)) 

x2, data2=np.loadtxt("risultati/FitMedioCircle.dat", unpack=True, usecols = (0,1)) 


pylab.plot(x1, data1, '-b', label='Fitness')
pylab.plot(x2, data2, '-r', label='Fitness medio')
pylab.legend(loc='upper right')
plt.title('Cities on a circumference')
plt.xlabel('#generazione')
plt.ylabel('L')
plt.grid(True)

plt.show()

```

```python
x3, y3=np.loadtxt("risultati/BestPath1.dat", unpack=True, usecols = (0,1)) 
plt.plot(x3,y3, marker='o', color='blue')

x_start=-0.754701 
y_start=0.430426 

plt.plot(x_start, y_start, color="red", marker = "o")

plt.title('Circumference: BestPath')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1.01, 1.01)
plt.ylim(-1.01, 1.01)

plt.show()
```
