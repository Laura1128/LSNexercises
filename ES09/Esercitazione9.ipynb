### The Traveling Salesman Problem

### Exercise 09.1

Il problema del commesso viaggiatore consiste nel trovare il percorso più breve per visitare N città. Il percorso deve iniziare in una città fissata e concludersi nella stessa città.

Nel codice sviluppato per risolvere tale problema il percorso è rappresentato da una matrice Nx2, dove N=32 è il numero di città da visitare. Ogni riga contiene le coordinate della città, mentre l'indice di colonna individua l'ordine con il quale vengono visitate.

L'algoritmo genetico sviluppato per questa esercitazione crea una popolazione di M=100 individui a partire dal primo generato, chiamando 4 operatori di mutazione che intervengono con una probabilità del 30%. Dopodichè segue i seguenti step:
- ordina la generazione in base al fitness, calcolato misurando la distanza totale percorsa $L^{(1)}$
- seleziona due individui con un algoritmo che privilegia gli individui con fitness migliore: $j=N\times int(r^2)$ dove $r$ è un numnero casuale $r\in[0,1)$
- chiama uno alla volta i quattro operatori di mutazione, che intervengono con una probabilità del 10% 
- viene chiamato l'operatore di crossover, che agisce con l'80% di probabilità 

Il processo viene reiterato fino a produrre 92 individui, che si aggiungeranno agli 8 migliori della generazione precedente per costituire la nuova generazione.

La successione di step descritta viene ripetuta per creare le generazioni successive, per un numero di volte sufficiente a trovare un minimo di $L^{(1)}$.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math
import pylab

x1, data1=np.loadtxt("risultati/BestFitCircle.dat", unpack=True, usecols = (0,1)) 

x2, data2=np.loadtxt("risultati/FitMedioCircle.dat", unpack=True, usecols = (0,1)) 


pylab.plot(x1, data1, '-b', label='Fitness')
pylab.plot(x2, data2, '-r', label='Fitness medio')
pylab.legend(loc='upper right')
plt.title('Cities on a circumference')
plt.xlabel('#generazione')
plt.ylabel('L')
plt.grid(True)

plt.show()

```
