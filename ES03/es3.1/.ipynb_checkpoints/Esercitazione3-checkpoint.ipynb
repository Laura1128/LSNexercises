### Exercise 03.1 <span style="color:red"> Plain vanilla option pricing </span>

Simulo il prezzo finale di un'opzione Call e un'opzione Put in due modi:
- campionando direttamente il valore finale $S(T)$
- dividendo l'intervallo temporale $[0,T]$ in $10^2$ sottointervalli e costruendo la traiettoria del valore dell'opzione $S(0)$, $S(t_1)$, $S(t_2)$, ... fino a $S(t_{100}=T=1)$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

M=10000
N=100                 # Number of blocks
L=int(M/N)            # Number of throws in each block
x = np.arange(N) 

data=np.loadtxt("Calldiretta.txt")
error=np.loadtxt("IncertezzeC1.txt")  
datad=np.loadtxt("Calldiscr.txt")
errord=np.loadtxt("IncertezzeC2.txt")  
plt.errorbar(x,data, yerr=error)
plt.axhline(y=14.975790778311286, xmin=0., xmax=100, color='r')
plt.title('Call diretto')
plt.xlabel('#blocchi')
plt.ylabel('$C[S(0),0]$')
plt.grid(True)
plt.figure()
plt.errorbar(x,datad, yerr=errord)
plt.axhline(y=14.975790778311286, xmin=0., xmax=100, color='r')
plt.title('Call discretizzato')
plt.xlabel('#blocchi')
plt.ylabel('$C[S(0),0]$')
plt.grid(True)
plt.show()
```
