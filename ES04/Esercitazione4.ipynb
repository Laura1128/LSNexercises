### Exercise 04.1

Di seguito vengono mostrati i valori istantanei di temperatura, energia cinetica, energia potenziale ed energia totale calcolati con sei simulazioni successive di un sistema di 108 particelle. Tutte le quantità sono espresse in unità ridotte.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


N=600               
L=10            
x = np.arange(N) 

dataT=np.loadtxt("es04.1/output_temp.dat")  
dataEkin=np.loadtxt("es04.1/output_ekin.dat") 
dataEpot=np.loadtxt("es04.1/output_epot.dat")  
dataEtot=np.loadtxt("es04.1/output_etot.dat")  
x*=L

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 8))
fig.suptitle('Equilibrazione del sistema')
ax1.plot(x,dataT)
ax1.axhline(y=1.4, xmin=0., xmax=6000, color='r')
ax1.set(xlabel='time step', ylabel='T*')
ax1.grid(True)

ax2.plot(x,dataEkin)
ax2.set(xlabel='time step', ylabel='$E_{kin}*/N$')
ax2.grid(True)

ax3.plot(x,dataEpot)
ax3.set(xlabel='time step', ylabel='$E_{pot}*/N$')
ax3.grid(True)

ax4.plot(x,dataEtot)
ax4.set(xlabel='time step', ylabel='$E_{tot}*/N$')
ax4.grid(True)
plt.show()
```

Superate le prime tre simulazioni, durante le quali si osservano oscillazioni molto ampie, la temperatura si assesta attorno al valore target $T=1.4$. Le ripartenze costringono il sistema a vivere alla temperatura desiderata, portandolo cioè a raggiungere l'equilibrio. Una volta equilibrato, il sistema mostra proprietà costanti nel tempo.

### Exercise 04.2

Dopo aver equilibrato il sistema, è possibile fare una simulazione più lunga durante la quale vengono misurati i valori medi di temperatura, energia cinetica, energia potenziale ed energia totale. Di seguito sono riportati tali valori con le relative incertezze in funzione del numero di blocchi. 

```python
N=100  

x = np.arange(N) 

dataT, errorT=np.loadtxt("es04.2/ave_temp.dat", unpack=True, usecols = (2,3))  
dataEkin, errorEkin =np.loadtxt("es04.2/ave_ekin.dat", unpack=True, usecols = (2,3)) 
dataEpot, errorEpot=np.loadtxt("es04.2/ave_epot.dat", unpack=True, usecols = (2,3))  
dataEtot, errorEtot=np.loadtxt("es04.2/ave_etot.dat", unpack=True, usecols = (2,3)) 

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 8))
fig.suptitle('Equilibrazione del sistema')
ax1.errorbar(x,dataT, errorT)
ax1.set(xlabel='#block', ylabel='<T*>')
ax1.grid(True)

ax2.errorbar(x,dataEkin, errorEkin)
ax2.set(xlabel='#block', ylabel='<$E_{kin}*/N$>')
ax2.grid(True)

ax3.errorbar(x,dataEpot, errorEpot)
ax3.set(xlabel='#block', ylabel='<$E_{pot}*/N$>')
ax3.grid(True)

ax4.errorbar(x,dataEtot, errorEtot)
ax4.set(xlabel='#block', ylabel='<$E_{tot}*/N$>')
ax4.grid(True)
plt.show()
```

### Exercise 04.3

Vengono di seguito mostrati i risultati delle simulazioni di un sistema di particelle di Argon ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu)

1. fase solida: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math


N=400                

x = np.arange(N) 
data=np.loadtxt("es04.3/solid/output_temp.dat")
x*=10

plt.title('Temperatura in fase di equilibrazione')
plt.plot(x,data)
plt.axhline(y=0.8, xmin=0., xmax=4000, color='r')
plt.xlabel('time step')
plt.ylabel('T')
plt.grid(True)
plt.show()
```

La temperatura inizialmente crolla per poi assestarsi intorno al valore target. Questo è dovuto al fatto che il sistema parte da una configurazione cristallina perfetta, dunque vicino al minimo del potenziale. Il movimento degli atomi che si ha facendo evolvere il sistema provoca un aumento dell'energia potenziale e una conseguente diminuzione dell'energia cinetica, che è direttamente proporzionale alla temperatura. In questo modo l'energia totale si conserva.

Una volta che il sistema è equilibrato, è possibile misurare le quantità termidinamiche.

```python
k_b= 1.38054*10**-23

x=np.arange(100)

dataT, errorT=np.loadtxt("es04.3/solid/ave_temp.dat", unpack=True, usecols = (2,3))  
dataEkin, errorEkin=np.loadtxt("es04.3/solid/ave_ekin.dat", unpack=True, usecols = (2,3)) 
dataEpot, errorEpot=np.loadtxt("es04.3/solid/ave_epot.dat", unpack=True, usecols = (2,3))  
dataEtot, errorEtot=np.loadtxt("es04.3/solid/ave_etot.dat", unpack=True, usecols = (2,3)) 
dataT*=120
dataEkin*=120*k_b
dataEpot*=(120*k_b)
dataEtot*=(120*k_b)
errorT*=120
errorEkin*=120*k_b
errorEpot*=(120*k_b)
errorEtot*=(120*k_b)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 8))
fig.suptitle('Fase solida')
ax1.errorbar(x,dataT, errorT)
ax1.set(xlabel='#block', ylabel='<T> [K]')
ax1.grid(True)

ax2.errorbar(x,dataEkin, errorEkin)
ax2.set(xlabel='#block', ylabel='<$E_{kin}/N$> [J]')
ax2.grid(True)

ax3.errorbar(x,dataEpot, errorEpot)
ax3.set(xlabel='#block', ylabel='<$E_{pot}/N$> [J]')
ax3.grid(True)

ax4.errorbar(x,dataEtot, errorEtot)
ax4.set(xlabel='#block', ylabel='<$E_{tot}/N$> [J]')
ax4.grid(True)
plt.show()
```
